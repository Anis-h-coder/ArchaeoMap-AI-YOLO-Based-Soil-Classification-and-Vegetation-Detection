```python
import ultralytics
ultralytics.checks()
```

```python
data_yaml_content = """
train: /content/Soil_Erosion_Dataset/train/images
val: /content/Soil_Erosion_Dataset/valid/images
test: /content/Soil_Erosion_Dataset/test/images

nc: 3
names: ['soil', 'vegetation', 'rock']
"""

with open('/content/Soil_Erosion_Dataset/data.yaml', 'w') as f:
    f.write(data_yaml_content)

print("✅ data.yaml created successfully!")
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data='/content/Soil_Erosion_Dataset/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    name="soil_erosion_seg"
)
```
